```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("/Users/nannu/Desktop/ML journey/pandasworks/company_sale_data/company_sales_data (2).csv")
df.columns
```

```python
monthly_summary=df.groupby("month_number")["total_profit"].sum()
monthly_summary
```

```python
plt.plot(monthly_summary.index,monthly_summary.values)
```

```python
units_sold=df.groupby("month_number")["total_units"].sum()
units_sold
```

```python
plt.plot(units_sold.index,units_sold.values,marker="o",color="red")
```

```python
products=['facecream', 'facewash', 'toothpaste', 'bathingsoap','shampoo', 'moisturizer']
for p in products:
    plt.plot(df["month_number"],df[p],label=p)
plt.legend()    
```
